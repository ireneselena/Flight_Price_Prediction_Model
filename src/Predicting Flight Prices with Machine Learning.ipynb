#### Checking for Outliers in Numerical Columns

In this dataset we have three numerical columns, namely: `duration`, `days_left`, `price`. We are going to check for any outliers and abnormalities within the value of those columns.

```python
# Create dataframe for numerical coloms
numeric_col = df.select_dtypes(include=['number'])
```

```python
# Create an empty list to store data
outliers_info = []

for col in numeric_col.columns:
    # Calculate Q1, Q3, and IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # Calculate upper and lower limits
    lower_bound = max(Q1 - 1.5 * IQR, 0)  # Set lower_bound to 0 if it's less than 0
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers_condition = (df[col] < lower_bound) | (df[col] > upper_bound)
    outliers = df[outliers_condition]

    # Add iteration results to outliers_info
    outliers_info.append([col, round(lower_bound, 2), round(upper_bound, 2), outliers_condition.sum()])

# Turn outliers_info into a dataframe
df_outliers = pd.DataFrame(outliers_info, columns=['Columns', 'Lower Bound', 'Upper Bound', 'Numbers of Outliers']).set_index('Columns').reset_index()
df_outliers
```

#### Normality Test

```python
# Create an empty list to store data
distribution = []

# Looping for each column in numeric_col
for col in numeric_col.columns:
    # Perform normality test with Shapiro-Wilk
    stat, p_value = shapiro(df[col])
    
    # Round the p-value to 3 decimal places
    p_value = round(p_value, 3)
    
    # Determine distribution based on p-value
    dist = "Normally Distributed" if p_value > 0.05 else "Not Normally Distributed"
    
    # Save the results to a list
    distribution.append([col, stat, p_value, dist])

# Create a DataFrame from a list
df_distribution = pd.DataFrame(distribution, columns=['Column', 'Statistic', 'P-Value', 'Distribution'])
df_distribution
```

Based on this result, we can see that none of the numerical columns are normally distributed.

<h4 style='text-align:left; color:darkcyan'><b> Metric Evaluation Used </b></h4>

In this experiment, the evaluation metrics used are Mean Absolute Error (MAE) and Mean Absolute Percentage Error (MAPE).

- **Mean Absolute Error (MAE)**

    **Mean Absolute Error (MAE)** is an evaluation metric used to measure how accurately a regression model makes predictions. MAE provides an average of the absolute errors between the values predicted by the model and the actual values from the data.

    Formula for Calculating MAE:
    $$ 
    \text{MAE} = \frac{1}{n} \sum_{i=1}^{n} | y_i - \hat{y}_i | 
    $$

    where:
    - $y_i$ is the actual value (observed value).
    - $\hat{y}_i$ is the predicted value generated by the model.
    - $n$ is the total number of data points (number of observations).

    **Explanation of the MAE Formula**

    1. **Calculation of Absolute Error**: For each data point $i$, calculate the absolute error between the actual value $y_i$ and the predicted value $\hat{y}_i$. This is done using $| y_i - \hat{y}_i |$, which results in a positive or zero value.

    2. **Averaging the Error**: Sum all the absolute errors and divide by the total number of data points $n$ to get the average error.

- **Mean Absolute Percentage Error (MAPE)**

    **Mean Absolute Percentage Error (MAPE)** is an evaluation metric that measures how large the model's errors are in terms of percentage relative to the actual values. This provides an interpretation of relative errors that is easier to understand.

    Formula for Calculating MAPE:
    $$
    \text{MAPE} = \frac{1}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right| \times 100\%
    $$
    
    where:
    - $y_i$ is the actual value (observed value).
    - $\hat{y}_i$ is the predicted value generated by the model.
    - $n$ is the total number of data points (number of observations).

    **Explanation of the MAPE Formula**

    1. **Calculation of Relative Error**: For each data point $i$, calculate the relative error between the actual value $y_i$ and the predicted value $\hat{y}_i$. This is done using $\left| \frac{y_i - \hat{y}_i}{y_i} \right|$, which results in a positive value.

    2. **Averaging the Relative Error**: Sum all the relative errors and divide by the total number of data points $n$ to get the average relative error.

    3. **Conversion to Percentage**: Multiply the result by 100% to obtain the error in percentage terms.

---

### Why Use MAE and MAPE Over MSE and RMSE?

**MAE** and **MAPE** are often preferred over **Mean Squared Error (MSE)** and **Root Mean Squared Error (RMSE)** for the following reasons:

1. **Interpretability**: 
   - MAE and MAPE provide errors in the same units as the target variable, which makes them easier to interpret. MAPE, in particular, gives the error in percentage terms, which is intuitive and understandable for stakeholders.
   
2. **Robustness to Outliers**:
   - MAE and MAPE are less sensitive to outliers compared to MSE and RMSE. Since MSE and RMSE square the errors, large errors (due to outliers) have a disproportionate impact, which can skew the evaluation metric. MAE and MAPE treat all errors equally, which can provide a more balanced view of model performance.

3. **Applicability**:
   - MAPE is especially useful in cases where the scale of the data can vary, as it normalizes the error relative to the actual value. This makes MAPE applicable across datasets with different scales.

4. **Alignment with Business Goals**:
   - In many business contexts, understanding the absolute error or the percentage error is more relevant than understanding the squared error. MAE and MAPE directly answer the question of "How far off, on average, are the predictions from the actual values?" which aligns closely with business objectives.

However, itâ€™s important to note that MAPE has limitations when the actual values are close to zero, as it can result in very large or undefined percentages. Therefore, it is essential to choose the evaluation metric that best suits the context of the problem.

```python
# Create Dict for Metric Evaluation
scoring = {
    'mae': 'neg_mean_absolute_error',
    'mape': 'neg_mean_absolute_percentage_error'
}

# Create function for Evaluation Model
def evaluate_models1(models1, pipeline, X_train, y_train, scoring):
    # Dict to save evaluation results
    results = {
        'Model': [],
        'mean_mae_train': [],
        'std_mae_train': [],
        'mean_mae_test': [],
        'std_mae_test': [],
        'mean_mape_train': [],
        'std_mape_train': [],
        'mean_mape_test': [],
        'std_mape_test': []
    }

    # Evaluate each model with the given pipeline
    for model, model_name in models1.items():
        # Create a pipeline for the model
        pipeline = Pipeline(steps=[
            ('preprocessor', preprocessor_1),
            ('model', model)])
        
        # Cross-validation to evaluate pipelines with different models
        cv_score = cross_validate(
            estimator=pipeline,
            X=X_train,
            y=y_train,
            scoring=scoring,
            cv=10,
            return_train_score=True
        )
        
        # Collect evaluation results for each metric
        results['Model'].append(model_name)
        
        # Mean and std for MAE
        results['mean_mae_train'].append(-np.mean(cv_score['train_mae']))
        results['std_mae_train'].append(np.std(cv_score['train_mae']))
        results['mean_mae_test'].append(-np.mean(cv_score['test_mae']))
        results['std_mae_test'].append(np.std(cv_score['test_mae']))

        # Mean and std for MAPE
        results['mean_mape_train'].append(-np.mean(cv_score['train_mape']))
        results['std_mape_train'].append(np.std(cv_score['train_mape']))
        results['mean_mape_test'].append(-np.mean(cv_score['test_mape']))
        results['std_mape_test'].append(np.std(cv_score['test_mape']))

    # Create a DataFrame from the evaluation results
    results_df = pd.DataFrame(results)
    
    return results_df

# Create function for Evaluation Model
def evaluate_models2(models2, pipeline, X_train, y_train, scoring):
    # Dict to save evaluation results
    results = {
        'Model': [],
        'mean_mae_train': [],
        'std_mae_train': [],
        'mean_mae_test': [],
        'std_mae_test': [],
        'mean_mape_train': [],
        'std_mape_train': [],
        'mean_mape_test': [],
        'std_mape_test': []
    }

    # Evaluate each model with the given pipeline
    for model, model_name in models2.items():
        # Create a pipeline for the model
        pipeline = Pipeline(steps=[
            ('preprocessor', preprocessor_2),
            ('model', model)])
        
        # Cross-validation to evaluate pipelines with different models
        cv_score = cross_validate(
            estimator=pipeline,
            X=X_train,
            y=y_train,
            scoring=scoring,
            cv=10,
            return_train_score=True
        )
        
        # Collect evaluation results for each metric
        results['Model'].append(model_name)
        
        # Mean and std for MAE
        results['mean_mae_train'].append(-np.mean(cv_score['train_mae']))
        results['std_mae_train'].append(np.std(cv_score['train_mae']))
        results['mean_mae_test'].append(-np.mean(cv_score['test_mae']))
        results['std_mae_test'].append(np.std(cv_score['test_mae']))

        # Mean and std for MAPE
        results['mean_mape_train'].append(-np.mean(cv_score['train_mape']))
        results['std_mape_train'].append(np.std(cv_score['train_mape']))
        results['mean_mape_test'].append(-np.mean(cv_score['test_mape']))
        results['std_mape_test'].append(np.std(cv_score['test_mape']))

    # Create a DataFrame from the evaluation results
    results_df = pd.DataFrame(results)
    
    return results_df
```

#### Create Function to Randomized Search CV

```python
def randomized_search(pipeline, parameters, scoring, X_train, y_train):
    # Menentukan RandomizedSearchCV dengan pipeline dan parameter distribusi
    randomized_search = RandomizedSearchCV(
        estimator=pipeline,
        param_distributions=parameters,
        n_iter=5,
        refit='mape',
        scoring=scoring,
        cv=5,
        random_state=42,
        return_train_score=True
    )
    
    # Melatih RandomizedSearchCV dengan data pelatihan
    randomized_search.fit(X_train, y_train)
    
    return randomized_search
```

#### Create Function to View Result of cv_result_

```python
# Create list metric evaluation
metrics = ['mae', 'mape']

# Create function
def cv_result(cv_results, metrics):
    # Konversi hasil cv_results ke DataFrame
    cv_results_df = pd.DataFrame(cv_results)
    
    results = {}
    
    for metric in metrics:
        train_mean_col = f'mean_train_{metric}'
        train_std_col = f'std_train_{metric}'
        test_mean_col = f'mean_test_{metric}'
        test_std_col = f'std_test_{metric}'
        
        # Mean and std for metrik train
        results[f'mean_{metric}_train'] = [-cv_results_df[train_mean_col].mean()] if metric in ['mae', 'mape'] else [cv_results_df[train_mean_col].mean()]
        results[f'std_{metric}_train'] = [cv_results_df[train_std_col].mean()]
        
        # Mean and std for metrik test
        results[f'mean_{metric}_test'] = [-cv_results_df[test_mean_col].mean()] if metric in ['mae', 'mape'] else [cv_results_df[test_mean_col].mean()]
        results[f'std_{metric}_test'] = [cv_results_df[test_std_col].mean()]

    return pd.DataFrame(results)
```

<h4 style='text-align:left; color:darkcyan'><b> How the Final Model Works </b></h4>

**The Final Model for this Customer Lifetime Value Prediction Case uses the Extra Trees Regressor Algorithm with the following parameters:**

- **`n_estimators=400`**: The number of trees in the forest. This parameter controls the number of trees that are built in the model. A higher number of trees generally improves model performance but also increases training time.

- **`criterion='squared_error'`**: The function to measure the quality of a split. The squared error is minimized, which is appropriate for regression tasks where the target variable is continuous.

- **`max_depth=15`**: The maximum depth of the tree. This parameter prevents the trees from growing too deep and overfitting by controlling the maximum number of levels in each decision tree.

- **`min_samples_split=4`**: The minimum number of samples required to split an internal node. This prevents the model from splitting nodes that are too small, which can lead to overfitting.

- **`min_samples_leaf=3`**: The minimum number of samples required to be at a leaf node. This ensures that the leaves have enough samples to make reliable predictions.

- **`max_features='auto'`**: The number of features to consider when looking for the best split. 'Auto' means that all features are considered.

- **`random_state=42`**: Sets the seed for random number generation, ensuring that the results are reproducible.

**Then, the Extra Trees Regressor Algorithm is defined as the variable `final_model_evl`, which is used within the TransformedTargetRegressor:**

- **`TransformedTargetRegressor`**: Wraps the Extra Trees Regressor model with target transformation. The target variable is log-transformed using `np.log` before the model is trained, and predictions are returned to the original scale using `np.exp`. This is useful when the target distribution is skewed, as log transformation can help the model handle the target variable more effectively.

**Pipeline**

- **`Pipeline`**: Combines several data preprocessing and model training steps into a single object:
  - **`'preprocessor'`**: This step includes data preprocessing such as normalization using `RobustScaler()`, encoding (Ordinal Encoding, OneHot Encoding), outlier detection, and imputation applied to the data before fitting the model.
  - **`'model'`**: Applies the Extra Trees Regressor model wrapped by `TransformedTargetRegressor`.

**Fitting the Data**

- **`final.fit(X_train, y_train)`**: Trains the pipeline on the training data (X_train and y_train). The pipeline applies preprocessing to the data and then trains the Extra Trees Regressor with the processed data and the transformed target.

**How the Extra Trees Regressor Model Works**

- **Preprocessing**:

  The data (X_train) is processed by the preprocessor, which may include normalization, encoding, or other feature transformations, ensuring that the data is in the optimal form for model training.

- **Target Transformation**:

  The target variable (y_train) is transformed using the log function (`np.log`) before training the Extra Trees Regressor model. This can improve model performance by addressing skewness in the target distribution. After training, the predictions are transformed back to the original scale using the exponential function (`np.exp`).

- **Model Training**:

  The Extra Trees Regressor is trained with the processed data (X_train) and the transformed target. The model constructs multiple trees, each built on different random subsets of the data and features, to create an ensemble that averages out predictions, reducing variance and improving model robustness.

- **Prediction**:

  After the model is trained, predictions can be made on new data. The predictions, initially in the transformed scale (log), are returned to the original scale using the exponential function (`np.exp`).

- **Evaluation**:

  The model is evaluated using regression metrics, such as Mean Absolute Error (MAE), on the test data to assess how well the model predicts the target variable.

This pipeline ensures that the data is preprocessed consistently, the target variable is handled appropriately, and the model is trained and evaluated effectively.
